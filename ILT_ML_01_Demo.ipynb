```python
%%writefile /content/bangkit2023-ilt-ml-01/AV/sum1_to_n.py
def sum1_to_n(n):
  """
  Calculate the sum of an arithmetic sequence from 1 to n.
  
  Args:
    n (int): last term in the sequence.
  
  Returns:
    The sum of 1 + 2 + ... + n.
  """
  return 0.5*n*(n + 1)
```

```python
%%writefile /content/bangkit2023-ilt-ml-01/AV/greeting.py
def greeting(name):
  """
  Say hello to the specified name, e.g., user.

  Args:
    name (str): name of the user.
  
  Returns:
    Hello, name.
  """
  return 'Hello, ' + name
```
